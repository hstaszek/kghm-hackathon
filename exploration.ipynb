```python
import numpy as np

with open('labels.txt', 'r') as f:
    labels = f.readlines()

stripped = list()
for lab in labels:
    _id, name = lab.split('\t')
    stripped.append((_id, name[:len(name)-1]))


stripped = np.array(stripped)
stripped[0:5]
```

```python
sns.heatmap(raw_df[stripped.take([10, 24, 21, 12, 11, 15, 17, 131, 132])].corr(), annot=True)
```

```python
to_plot = np.array([132,134,143,144,145,146,133,135,136]) - 2
sns.heatmap(raw_df[stripped.take(to_plot)].corr(), annot=True)
```

```python
to_plot = ['LMAM_HC201A_PLKL90---_TPG', 'LMAM_HC201A_PL-------_TPS', 'LMAM_TM211-_RDFQ01---_TFI', 'LMAM_RZM211_NDLT01---_TLI']
plot_legend = ['Uziarnienie dla HC-201A.KL90', 'Średnie uziarnienie dla HC-201A', 'Przepływ rudy dla TM-211', 'Poziom nadawy dla RZM-211']
for feat in to_plot:
    plt.plot(raw_df.index[0:60], raw_df[feat].iloc[0:60])
plt.legend(plot_legend)
plt.tight_layout()
plt.xlabel('Czas (min)')
#plt.ylim((0,80))
```
